```python
from tabula import read_pdf
import requests
import pandas as pd
import datetime
import numpy as np
import plotly.graph_objects as go
import plotly.express as px
import os
import warnings
import pickle
base_directory = "Stats"
columns= ['Country','Total_confirmed_cases_day',
'Total_new_cases_day','Total_deaths_day','Total_new_deaths_day','classification_day',
'Days_after_last_reported_case_day']
```

```python
def parse_reports(file_path,pages=[2,3,4],column_idx = 0):
    df = read_pdf(file_path,pages =pages)
    data_1,data_2,data_3 = df[0],df[1],df[2]
    data_1.drop(1,inplace=True)
    data_1.dropna(inplace=True)
    data_2.dropna(inplace =True)
    print(data_2.head())
    data_3.dropna(inplace=True)
    dataframeOne = pd.DataFrame.from_records(data_1.values,columns = columns)
    dataframeTwo = pd.DataFrame.from_records(data_2.values,columns= columns)
    dataframeThree = pd.DataFrame.from_records(data_3.values,columns=columns)
    dataframe = pd.concat([dataframeOne,dataframeTwo,dataframeThree])
    dataframe.drop("classification_day".format(column_idx),1,inplace=True)
    dataframe['day'] = [column_idx]*len(dataframe)
    return dataframe
    
```

```python
datasets = [] 
datasets.append(parse_reports(os.path.join("Stats","report_60.pdf"),column_idx = 0))
datasets.append(parse_reports(os.path.join("Stats","report_61.pdf"),column_idx = 1))
datasets.append(parse_reports(os.path.join("Stats","report_62.pdf"),column_idx = 2))
datasets.append(parse_reports(os.path.join("Stats","report_63.pdf"),column_idx = 3))
datasets.append(parse_reports(os.path.join("Stats","report_64.pdf"),pages=[3,4,5],column_idx = 4))
datasets.append(parse_reports(os.path.join("Stats","report_65.pdf"),pages=[4,5,6],column_idx = 5))
datasets.append(parse_reports(os.path.join("Stats","report_66.pdf"),pages=[3,4,5],column_idx = 6))
datasets.append(parse_reports(os.path.join("Stats","report_67.pdf"),pages=[4,5,6],column_idx = 7))
datasets.append(parse_reports(os.path.join("Stats","report_68.pdf"),column_idx = 8))
datasets.append(parse_reports(os.path.join("Stats","report_69.pdf"),column_idx = 9))
datasets.append(parse_reports(os.path.join("Stats","report_70.pdf"),column_idx = 10))
datasets.append(parse_reports(os.path.join("Stats","report_71.pdf"),pages=[3,4,5],column_idx = 11))
datasets.append(parse_reports(os.path.join("Stats","report_72.pdf"),pages=[4,5,6],column_idx = 12))
```

```python
dataset = datasets[0]
for d in range(1,len(datasets)):
    dataset = pd.concat([dataset,datasets[d]])
dataset.head()
dataset.to_csv('Stats/statsForLast14days.csv',sep='\t',encoding='utf-8')
```
